```python
import csv

f = open('/content/drive/MyDrive/DATA/accident.csv', encoding = 'cp949')

data = csv.reader(f)

next(data) # 한 행 건너뛰기

days_sum = [0]*7  # 요일 별 합을 저장하는 리스트
days = []
i = 0
k = 0

for row in data:
  row[2 : ] = map(int, row[2: ])
  days_sum[i%7] += row[2]
  i += 1

  if k < 7:
    days.append(row[1])
    k += 1

print(days)
print(days_sum)

plt.title("요일 별 교토사고 발생 건수")
plt.bar(days, days_sum)
plt.show()
```

```python
import csv

f = open('/content/drive/MyDrive/DATA/accident.csv', encoding = 'cp949')

data = csv.reader(f)

next(data) # 한 행 건너뛰기

i = 0

road_type = []
number = []

print("0:일 / 1:월 / 2:화 / 3:수 / 4:목 / 5:금 / 6:토")
day = int(input('요일을 숫자로 입력하세요 : '))
for row in data:
  row[2 : ] = map(int, row[2: ])
  if i % 7 == day:
    road_type.append(row[0])
    number.append(row[2])

  i += 1

for i in range(len(road_type)):
  print(road_type[i], ":", number[i])

# plt.title("요일 별 교통사고 발생 건수")
# plt.bar(road_type, number)
# plt.show()


plt.title("요일 별 교통사고 발생 건수")
plt.axis("equal")
plt.pie(number, labels = road_type, autopct = "%.1f%%")
plt.show()
```

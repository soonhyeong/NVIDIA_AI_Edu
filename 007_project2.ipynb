```python
import csv

f = open('/content/drive/MyDrive/DATA/SNS.csv', encoding = 'utf-8')

data = csv.reader(f)

next(data)
next(data)

facebook = []
year = []
insta = []

for row in data:
  facebook.append(float(row[3]))
  insta.append(float(row[5]))
  year.append(int(row[0]))

print(facebook, year)

plt.figure(figsize=(8, 4))
plt.plot(year, facebook, label = 'facebook', color = 'blue')
plt.plot(year, insta,  label = 'instagram', color = 'pink')
plt.legend() # label 표시

plt.xticks(year, rotation=45)  # 연도가 정수로 표현되게

plt.show()

f.close()
```

# 1. numpy — 숫자 배열 다루기

## 1-1. numpy란 무엇인가
numpy(넘파이)는 많은 숫자를 한꺼번에 빠르게 계산하기 위한 라이브러리입니다. 핵심 자료구조는 **배열(ndarray)** 입니다.

리스트와 배열의 차이
- 리스트는 서로 다른 자료형을 섞어 담을 수 있지만, 배열은 **모든 원소가 같은 자료형**입니다.
- 배열은 원소 전체에 대한 계산(더하기, 곱하기 등)을 **반복문 없이 한 번에** 처리합니다. 이것을 벡터화(vectorization)라고 하며 훨씬 빠릅니다.

관례적으로 `import numpy as np` 로 불러옵니다.

````python
# 리스트 vs numpy 배열
py_list = [1, 2, 3, 4]
arr = np.array([1, 2, 3, 4])
print("리스트:", py_list)
print("배열:", arr)

# 배열은 반복문 없이 전체에 연산이 적용됩니다 (벡터화)
print("각 원소 x2:", arr * 2)
print("각 원소 +10:", arr + 10)
````

## 1-2. 배열 만들기
자주 쓰는 배열 생성 함수입니다.

| 함수 | 기능 |
| --- | --- |
| `np.array([...])` | 리스트로부터 배열 생성 |
| `np.arange(start, stop, step)` | 일정 간격의 정수 배열 (range와 유사) |
| `np.linspace(start, stop, n)` | start~stop을 n등분한 배열 |
| `np.zeros(n)` / `np.ones(n)` | 0 또는 1로 채운 배열 |
| `np.random.rand(n)` | 0~1 사이 난수 배열 |

````python
print(np.arange(0, 10, 2))       # 0부터 2씩 증가
print(np.linspace(0, 1, 5))      # 0~1을 5등분
print(np.zeros(3))               # [0. 0. 0.]
print(np.ones(3))                # [1. 1. 1.]
print(np.array([[1, 2, 3], [4, 5, 6]]))   # 2차원 배열
````

## 1-3. 배열의 속성
배열의 모양과 종류를 알려주는 속성입니다.

| 속성 | 의미 |
| --- | --- |
| `arr.shape` | 배열의 모양 (행, 열) |
| `arr.ndim` | 차원 수 |
| `arr.size` | 전체 원소 개수 |
| `arr.dtype` | 원소의 자료형 |

````python
a = np.array([[1, 2, 3], [4, 5, 6]])
print("shape:", a.shape)   # (2, 3)
print("ndim :", a.ndim)    # 2
print("size :", a.size)    # 6
print("dtype:", a.dtype)   # int64
````

## 1-4. 인덱싱, 슬라이싱, 조건 선택
리스트처럼 인덱싱/슬라이싱이 되고, 조건(불리언)으로 원하는 값만 골라낼 수도 있습니다.

````python
arr = np.array([10, 20, 30, 40, 50])
print(arr[0], arr[-1])       # 10 50
print(arr[1:4])              # [20 30 40]

# 조건으로 골라내기 (불리언 인덱싱)
print(arr[arr >= 30])        # [30 40 50]
print(arr > 25)              # [False False  True  True  True]
````

## 1-5. 배열 연산과 브로드캐스팅
배열끼리의 연산은 **같은 위치의 원소끼리** 계산됩니다. 배열과 숫자(스칼라)의 연산은 모든 원소에 적용되는데, 이를 브로드캐스팅이라고 합니다.

````python
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])
print("a + b :", a + b)      # [11 22 33]
print("a * b :", a * b)      # [10 40 90]
print("a ** 2:", a ** 2)     # [1 4 9]
print("a + 100:", a + 100)   # 브로드캐스팅 [101 102 103]
````

## 1-6. 통계 함수
데이터 요약에 자주 쓰는 함수입니다. 2차원 배열에서는 `axis` 로 방향을 정합니다. (`axis=0` 열 방향, `axis=1` 행 방향)

| 함수 | 기능 |
| --- | --- |
| `arr.sum()` | 합계 |
| `arr.mean()` | 평균 |
| `arr.std()` | 표준편차 |
| `arr.min()` / `arr.max()` | 최소 / 최대 |
| `np.median(arr)` | 중앙값 |

````python
scores = np.array([85, 90, 78, 92, 88])
print("합계:", scores.sum())
print("평균:", scores.mean())
print("표준편차: %.2f" % scores.std())
print("최고/최저:", scores.max(), scores.min())
print("중앙값:", np.median(scores))
````

````python
# axis 예시: 과목별(열) 평균 vs 학생별(행) 평균
table = np.array([[90, 80, 70],    # 학생1: 국 영 수
                  [60, 100, 80]])  # 학생2
print("과목별 평균(axis=0):", table.mean(axis=0))   # 열 방향
print("학생별 평균(axis=1):", table.mean(axis=1))   # 행 방향
````

# 3. EDA — 탐색적 데이터 분석

## 3-1. EDA란 무엇인가
EDA(Exploratory Data Analysis, 탐색적 데이터 분석)는 본격적인 분석·모델링에 앞서 **데이터를 여러 각도로 살펴보며 특징과 패턴, 문제점을 파악하는 과정**입니다.

왜 필요할까요? 데이터를 모른 채 분석하면 잘못된 결론을 내리기 쉽습니다. EDA로 데이터의 분포, 관계, 이상치, 결측치를 먼저 이해하면 더 올바른 질문과 분석을 할 수 있습니다.

EDA의 일반적인 순서
1. 데이터 구조 파악: 크기, 열, 자료형 (`shape`, `info`)
2. 요약 통계 확인: 평균, 범위 등 (`describe`)
3. 결측치·이상치 점검
4. 한 변수의 분포 보기: 히스토그램, 박스플롯
5. 두 변수의 관계 보기: 산점도, 상관계수
6. 범주별 비교: 막대그래프

시각화에는 주로 matplotlib과 seaborn을 사용합니다.

## 3-2. 데이터 개요 파악
분석 대상을 정하고(펭귄 데이터), 크기와 요약을 먼저 봅니다.

````python
df = sns.load_dataset("penguins")
print("크기:", df.shape)
print("결측치 총합:", df.isnull().sum().sum())
df.describe()
````

## 3-3. 한 변수의 분포 보기
- 히스토그램(`histplot`): 값이 어떤 구간에 얼마나 몰려 있는지
- 박스플롯(`boxplot`): 중앙값, 사분위, 이상치를 한눈에

````python
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x="body_mass_g", bins=20)
plt.title("펭귄 몸무게 분포")
plt.show()
````

````python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x="species", y="body_mass_g")
plt.title("종별 몸무게 분포")
plt.show()
````

## 3-4. 범주형 변수 비교
- 개수 비교: `countplot`
- 평균 등 비교: `barplot` (기본은 평균과 오차막대)

````python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="species")
plt.title("종별 펭귄 수")
plt.show()
````

````python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x="species", y="bill_length_mm")
plt.title("종별 평균 부리 길이")
plt.show()
````

## 3-5. 두 변수의 관계 보기
- 산점도(`scatterplot`): 두 수치 변수의 관계
- 상관계수 히트맵(`heatmap`): 여러 수치 변수 간 관계를 색으로 (상관계수는 -1~1, 1에 가까울수록 강한 양의 관계)

````python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species")
plt.title("부리 길이 vs 부리 두께")
plt.show()
````

````python
# 수치 열들의 상관계수 히트맵
num_df = df.select_dtypes(include="number")
plt.figure(figsize=(6, 5))
sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("수치 변수 상관관계")
plt.show()
````

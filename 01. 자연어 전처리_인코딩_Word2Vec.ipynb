### 자연어 처리(Natural Language Processing, NLP)
- 인간의 언어를 컴퓨터가 이해하고 해석 및 생성을 할수 있게 만드는 인공지능 영역
- 언어번역, 감성분석, 챗봇, 텍스트 용약, 정보 추출 등 다양한 영역에서 사용

- 자연어 처리 단계
    1. 텍스트 전처리: 오타 수정, 불용어 제거, 토큰화(문장을 단어로 쪼개기)
    2. 임벤딩: 단어 또는 문장을 숫자 변경(벡터:Vector) => 의미가 비슷한 단어들을 숫자 상으로 가까운 위치에 정렬
    3. 모델링: 다양한 영역에 적용하기 위한 학습(LLM: Transformer 모델)

#### 2. 정제 및 정규화
- 불필요한 단어 제거, 유사 단어는 하나의 단어로 변경..

```python
from konlpy.tag import Okt
import re
```

```python
text = '자연어 처리... 대단해요!! 인공지능 NLP 학습하고 있음 ㅠㅠ, 너무 재밌어요 ㅋㅋㅋㅋㅋㅋㅋ #AI #열공'
```

- 정규 표현식을 이용한 한글, 영문, 숫자를 제외한 글자 제거

```python
clean_text = re.sub(r'[^가-힣ㄱ-ㅎㅏ-ㅣA-Za-z0-9 ]', '', text)
clean_text
```

- 토큰화(형태소 단위 추출)

```python
okt = Okt()
token = okt.morphs(clean_text)  # 입력값 기준 
print(token)
```

```python
token = okt.morphs(clean_text, stem=True) # 정규화: 있음 => 있다, 재밌어요 => 재밌다
print(token)
```

- 불용어(Stopword) 처리: 토큰화 된 단어들 중 조사, 접속사와 같은 불필요한 단어 제거

```python
stword = 'NLP 하고 있다 ㅠㅠ ㅋㅋㅋㅋㅋㅋㅋ'
stopword = stword.split()
stopword
```

```python
result_text=[]
for word in token:
    if word not in stopword:
        result_text.append(word)
        
print(result_text)
```

```python
result_text = [word for word in token if word not in stopword]
print(result_text)
```

#### 3. 정수 인코딩 & 원-핫 인코딩
- 정수 인코딩: 단어 사전(Vocabulary)을 만들고, 각 단어에 고유한 정수를 부여하는 방식
- 원-핫 인코딩: 단어 집합의 크기를 벡터의 차원으로 하고, 표현하고 싶은 단어의 인덱스에 1을, 나머지에는 0을 부여하는 방식

```python
raw_texts = [
    "인공지능 공부는 너무 재밌어요ㅋㅋㅋ",
    "데이터 분석은 정말 중요합니다.",
    "자연어 처리는 인공지능의 핵심 분야입니다."
]
```

```python
from konlpy.tag import Okt
import re
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer # 정수 인코딩
from tensorflow.keras.utils import to_categorical  # 원-핫 인코딩
```

- 데이터 전처리

```python
okt = Okt()
token_data =[]

st_text = "는 은 의 이다"
stopword = st_text.split()

for text in raw_texts:
    # 정제(한글과 공백을 제외한 나머지 제거)
    clean_text = re.sub(r'[^가-힣 ]', '', text)

    # 토큰화(정규화)
    tokens = okt.morphs(clean_text, stem=True)

    # 불용어 처리
    clean_tokens = [word for word in tokens if word not in stopword]
    
    token_data.append(clean_tokens)

token_data
```

- 정수 인코딩

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(token_data) # 단어 사전 생성

print(tokenizer.word_index)
```

```python
# 문장을 단어 사전을 이용해 스퀀스(숫자)로 변환
sequences_data = tokenizer.texts_to_sequences(token_data)
print(sequences_data)
```

- 원-핫 인코딩

```python
# 전체 단어의 크기(단어 갯수)
vocab_size = len(tokenizer.word_index) + 1

# 첫번째 데이터(문장)에 대한 원-핫 인코딩 수행
one_hot_vectors =  [to_categorical(seq, num_classes=vocab_size) for seq in sequences_data]
one_hot_vectors
```

#### 4. 워드 임베딩
- 단어를 벡터로 표현하는 방법
- 희소 표현: 원-핫 벡터(인코딩), 해당되는 단어에 1, 나머지는 0으로 표현, 유사도 평가 불가
    - 예: 인공지능 = [0, 1, 0, 0, ...., 0]  => 단어가 10000개 = 10000차원
- 밀집 표현: 비슷한 의미를 가진 단어들의 벡터 공간에서 가깝게 위치하도록 만드는 방법, 실수형 표현
    - 예: 인공지능 = [0.2, 1.8, 1.1, -2.1 .... ] => 128차원 정의 = 모든 단어를 128차원으로 표현
- 워드 임베딩 = 밀집 벡터 = 임베딩 벡터

##### 워드 임베딩 알고리즘
- Word2Vec: 가장 대표적인 알고리즘
    - 중심 단어로 주변 단어를 예측하거나(Skip-gram), 주변 단어로 중심 단어를 예측(CBOW)하며 학습
- GloVe: 윈도우 기반의 주변 단어 출현 빈도를 전체 텍스트 데이터 관점에서 통계적으로 반영
- FastText: 페이스북에서 개발
    - 단어를 '부분 단어(Subword)' 단위로 쪼개어 학습, 한국어에 강함

#### 워드 임베딩 예제

```python
from gensim.models import Word2Vec
from konlpy.tag import Okt
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

- 샘플 데이터

```python
texts = [
    "인공지능은 매우 흥미로운 분야입니다",
    "자연어 처리는 인공지능의 한 종류입니다",
    "딥러닝은 인공지능을 구현하는 기술입니다",
    "머신러닝은 데이터에서 패턴을 학습합니다",
    "자연어처리는 문장을 이해하고 생성합니다",
    "딥러닝 모델은 많은 데이터를 필요로 합니다"
]
```

- 토큰화 진행

```python
okt = Okt()
token_data = [ okt.morphs(txt) for txt in texts]

token_data
```

- Word2Vec 모델을 이용한 학습

```python
model = Word2Vec(sentences = token_data,
                 vector_size=128,   # 단어의 벡터 차원(128차원)
                 window=5,          # 중심단어 기준 앞뒤 5개의 단어까지 학습
                 min_count=1,       # 등장 빈도 => 1번 이상 등장한 모든 단어에 대해 학습
                 sg=0,              # 0=CBOW, 1=Skip-gram
                 workers=2)         # CPU 개수
```

- 단어 목록과 벡터 추출

```python
words = list(model.wv.index_to_key)   # 학습된 전체 단어
print(words)
print(len(words))
```

```python
vectors = np.array([model.wv[w] for w in words])
print(len(vectors[0]))
print(vectors.shape)
```

- PCA를 이용해 2차원 축소
    - PCA => 주성분 분석
    - 고차원의 데이터를 정보 손싱를 최소화 하면서 최원을 축소하는 방법

```python
pca = PCA(n_components=2)  # 128차원 => 2차원으로 차원축소
vectors_2d = pca.fit_transform(vectors)
print(vectors_2d.shape)    # (단어수, 2)
print(vectors_2d[0])
```

- vectors_2d 데이터를 이용한 시각화

```python
# 차트 한글 글꼴 지정
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

plt.figure(figsize=(10, 8))
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1])

for i, word in enumerate(words):
    plt.text(vectors_2d[i, 0], vectors_2d[i, 1], word)

plt.title('Word2Vec를 이용한 벡터 단어 시각화(PCA-2D)')
plt.xlabel('PCA1')
plt.ylabel('PCA1')
plt.grid(True)

plt.show()
```

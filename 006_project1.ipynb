```python
from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
text_content = uploaded[file_name].decode('utf-8')

# 5번 입력 후 출력 / 리스트 사용 안함
# for _ in range(5):
#    target_word = input()
#    word_count = text_content.count(target_word)
#    print("%s : %d번"%(target_word, word_count))


# 공백을 기준으로 문자 5개 입력 / 리스트 사용
target_word = input().split()
word_count = []
for i in range(len(target_word)):
   n = text_content.count(target_word[i])
   word_count.append(n)

for i in range(len(target_word)):
   print(target_word[i], word_count[i])


plt.figure(figsize=(8, 4))
plt.bar(target_word, word_count)  # x축에는 단어 리스트, y축에는 숫자 리스트를 넣음
plt.xticks(rotation=45)  # 단어 이름이 겹치지 않게 45도 회전
plt.show()
```

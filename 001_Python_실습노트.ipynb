### print() 함수
파이썬의 표준 출력 함수는 `print()` 입니다.

| 명령어 | 기능 |
| --- | --- |
| `print(A)` | A를 출력. A가 문자열이면 따옴표로 감쌉니다. |
| 여러 값 나열 | `+` 로 이으면 공백 없이, `,` 로 구분하면 값 사이에 공백 한 칸 |
| `end=` | `print(A, end='문자')` — 끝의 줄바꿈 대신 다른 문자로 (기본: 줄바꿈) |
| `sep=` | `print(A, B, sep='문자')` — 값 사이 공백 대신 다른 문자로 (기본: 공백) |

````python
# [프로그램 1-1] 출력 예제
print("Hello")
print("world")
````

````python
# [프로그램 1-2] end 옵션
print("Hello", end='!!!!!')
print("world")
````

````python
# [프로그램 1-3] sep 옵션
print("Hello", "world")
print("Hello", "world", sep=' good ')
````

````python
# [프로그램 1-4] 옵션 혼합
print("Hello", "world", "Hi", "python", sep='~', end='!')
````

## 🚀Let's Start — 변수·자료형·연산

### 변수 (Variable)
- **변수**는 '값'을 저장하는 공간에 붙인 **이름**입니다.
- 상자에 비유하면, 상자 이름을 `a` 라 하고 안에 `1` 을 넣으면 `a` 의 값은 `1`. 다시 `2` 를 넣으면 이전 값은 사라지고 `2` 가 됩니다.
- `=` 기호는 '같다'가 아니라 **'오른쪽 값을 왼쪽 변수에 저장하라'** 는 의미입니다.

````python
a = 1        # 변수 a에 1 저장
print(a)
a = 2        # a의 값을 2로 변경
print(a)
````

### 자료형 (Data type)
`type()` 함수로 자료형을 확인할 수 있습니다.

| 자료형 | 설명 | 예시 |
| --- | --- | --- |
| 정수 `int` | 양·음의 정수와 0 | `123, 0, -321` |
| 실수 `float` | 소수점 포함 | `123.45, -123.45` |
| 문자열 `str` | 문자들의 집합 | `"Hello World"` |
| 부울 `bool` | 참/거짓 | `True, False` |

````python
# [프로그램 2-1] 변수의 값과 자료형 출력
a = 123
b = "Hello World"
c = True
print(a, type(a))
print(b, type(b))
print(c, type(c))
````

### 포맷 코드로 문자열 포매팅
문자열 안에 값을 삽입하는 방법입니다. `%` 연산자와 문자를 조합하며, 삽입할 데이터는 따옴표 뒤에 `%` 와 함께 씁니다. (데이터가 여러 개면 괄호로 묶기)

| 포맷 코드 | 자료형 | 예시 |
| --- | --- | --- |
| `%d` | 정수 | `"%d은 정수이다." % 7` |
| `%f` | 실수 | `"원주율은 %f이다." % 3.14` |
| `%s` | 문자열 | `"나의 이름은 %s이다." % "홍길동"` |
| `%c` | 문자 | `"'%c' 등급입니다." % 'A'` |
| `%%` | % 문자 자체 | `"성공률 : 90%%"` |

````python
# 1. 정수 대입
print("%d은 정수이다." % 7)

# 2. 정수형 변수 대입
number = 3
print("%d은 정수이다." % number)

fa_num = 7
print("내가 가장 좋아하는 숫자는 %d 입니다."% fa_num)
````

````python
# 3. 2개 이상의 값
m = 12
d = 7
print("오늘은 %d월 %d일이다." % (m, d))

# 4. 실수 (기본 소수점 6자리)
pi = 3.42134234
print("%f" % pi)
````

````python
# 5. 소수점 자리수 지정
pi = 3.42134234
print("%.2f" % pi)      # 3.42

# 6. 자릿수 지정 (양수: 오른쪽 정렬, 음수: 왼쪽 정렬)
print("%-10s" % "hello")
print("%10.2f" % 3.141592653)

# 7. %s 는 어떤 값이든 문자열로 변환
print("나는 공책이 %s권 있다." % 3)
````

### format() 함수로 포매팅
포맷 코드 대신 `{ }` 중괄호와 인수 순서(0부터)를 사용합니다.

````python
# 1. 숫자 바로 대입
print("{0}은 정수이다.".format(7))

# 2. 2개 이상
name = "홍길동"
print("내 이름은 {0}이고, {1}월에 태어났다.".format(name, 5))
````

### 대입 연산자와 산술 연산자
- `=` 의 오른쪽에는 어떤 수식이라도 올 수 있고, 계산 결과가 왼쪽 변수에 저장됩니다.
- `a = b = 10` (같은 값 저장), `a, b = 1, 2` (여러 값 한 번에) 도 가능합니다.

| 기호 | 기능 | 예 (`a=10, b=3`) | 결과 |
| --- | --- | --- | --- |
| `+` `-` `*` | 덧셈·뺄셈·곱셈 | `a+b`, `a-b`, `a*b` | 13, 7, 30 |
| `/` | 나눗셈(실수) | `a/b` | 3.3333… |
| `//` | 몫 | `a//b` | 3 |
| `%` | 나머지 | `a%b` | 1 |
| `**` | 거듭제곱 | `a**b` | 1000 |

````python
a = 10
b = 2
print("%d %d %d" % (a + b, a - b, a * b))

print(7/2, 7//2, 7%2)          # 3.5 3 1

a = (100 + 96 + 80) // 3       # 괄호가 가장 먼저 계산됨
print("%d" % a)
````

### 형변환 (Type casting)

| --- | --- |
| `int(data)` | 정수형으로 |
| `float(data)` | 실수형으로 |
| `str(data)` | 문자열로 |

````python
# [프로그램 2-3] 간단한 계산기 1
a = input()
b = input()
a = int(a)             # 정수형으로 형변환
b = int(b)
print("%d + %d = %d" % (a, b, a+b))
````

### 인덱싱 (Indexing)
군집 자료형(문자열·리스트·튜플)의 값들은 **0부터** 시작하는 인덱스가 매겨져 있습니다.
`변수[인덱스]` 로 값에 접근합니다. **음수 인덱스**는 뒤에서부터(-1, -2, …) 셉니다.

`a = "Carpe Diem!"` 일 때: `a[0]` → `'C'`, `a[-1]` → `'!'`

````python
a = "Carpe Diem!"
print(a[0], a[4], a[-1], a[-2])

b = [10, 20, [30, 40], 50]
print(b[2], b[2][0], b[2][1])
````

### 슬라이싱 (Slicing)

### 문자열 내장 함수

| 함수 | 기능 |
| --- | --- |

| `a /= b` | `a = a / b` |

| `a.clear()` | 모든 쌍 삭제 |

````python
# [프로그램 6-5] 딕셔너리 함수 예제
a = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
print(a.keys())
print(a.values())
print(a.items())
a.clear()
print(a)
````
